```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
```

```python
from sklearn.compose import ColumnTransformer
numeric_cols = ['person_age', 'person_income', 'loan_amnt', 'loan_int_rate', 'cb_person_cred_hist_length', 'credit_score']
transform = ColumnTransformer(transformers=[
    ('tnf_gender_loan_intent', OneHotEncoder(sparse_output=False, drop=None, handle_unknown='ignore'), ['person_gender', 'loan_intent']),
    ('tnf_edu', OrdinalEncoder(categories=[['High School', 'Associate', 'Bachelor', 'Master', 'Doctorate']], handle_unknown='use_encoded_value', unknown_value=-1),['person_education']),
    ('tnf_ownership', OrdinalEncoder(categories=[['RENT', 'MORTGAGE', 'OWN']], handle_unknown='use_encoded_value', unknown_value=-1),['person_home_ownership']),
    ('tnf_previous_loan', OrdinalEncoder(categories=[['No', 'Yes']]), ['previous_loan_defaults_on_file']),
    ('scaler', StandardScaler(), numeric_cols)
], remainder='passthrough')
```

```python
y_train.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipeline = Pipeline(steps=[
    ('transform', transform),
    ('clf', RandomForestClassifier(random_state=42))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
predict = pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predict)
accuracy
```

```python
import joblib
joblib.dump(pipeline, 'loan_approval_pipeline.pkl')
```
